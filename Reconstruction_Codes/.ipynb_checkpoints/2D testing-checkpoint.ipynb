## Comparison with PyAbel

```python
# Analytical Abel: 
n = 100
r_max = 1
sigma = 0.5

ref = abel.tools.analytical.GaussianAnalytical(n, r_max, sigma,symmetric=False)
```

```python
m=500

frame = Image(np.zeros(n),r_max/n)
print("Image constructed")
circs = Isolines(0,r_max,n,m)
print("Classes constructed")
circs.sample()
print("Sampling done")
circs.triangulate()
print("Triangulation done")
I=circs.calculate_matrix_elements(frame)
inv = np.linalg.inv(I)
```

```python
I.sum()*frame.pixelsize-np.pi/2
```

```python
func=np.exp(-circs.R**2/sigma**2)
```

```python
plt.plot(ref.r,np.flip(I@func),marker='+',ms = 6,ls='None',label='Discrete Isoline Transform')
plt.plot(ref.r, ref.abel,color = 'black',alpha=0.6,label='Analytical transform, discretized')
plt.xlabel('y')
plt.ylabel('F(y)')
plt.legend()
plt.title('Forward Transform, N=50, M=100')
#plt.savefig('Gaussian_transformed')
plt.show()
```

```python
#reconion = abel.dasch.onion_peeling_transform(ref.abel, verbose=True, basis_dir=None, dr = ref.dr, direction='inverse')
plt.plot(ref.r,inv@np.flip(ref.abel),marker='+',ms=6,ls='None',label='Discrete Isoline Reconstruction')
plt.plot(ref.r,func,color='black',alpha=0.6,label='Original distribution')
plt.xlabel('r')
plt.ylabel('f(r)')
plt.legend()
plt.title('Backward transform, applied to analytical signal, N=100, M=500')
#plt.savefig('Gaussian_original_backward')
plt.show()
```

## Effect of noise

```python
func=np.exp(-circs.R**2/sigma**2)#+0.1*np.cos(5*circs.R)
```

```python
pure_signal = I@func
noise_ratio = 0.02
noise = pure_signal.std() * noise_ratio * np.random.normal(0,1,pure_signal.shape[0])
#noise = np.flip(np.sort(noise))

noisy_signal = pure_signal + noise
```

```python
plt.plot(ref.r,np.flip(noisy_signal),marker='+',ms=6,ls='None',label='Noisy signal')
plt.plot(ref.r,np.flip(pure_signal),color='black',alpha=0.6,label='Pure signal')
plt.xlabel('y')
plt.ylabel('F(y)')
plt.legend()
plt.title('Signal with 2% added white Gaussian noise')
plt.savefig('Noise')
plt.show()
```

```python
recon = abel.basex.basex_transform(np.flip(noisy_signal), verbose=True, basis_dir=None, dr = ref.dr, direction='inverse')
plt.plot(ref.r,recon,marker='+',ms=6,ls='None',label='BASEX reconstruction')
plt.plot(ref.r,inv@noisy_signal,marker='x',ms=6,ls='None',label='Discrete Isoline reconstruction')
plt.plot(ref.r,func,color='black',alpha=0.6,label='Original inverse')
plt.xlabel('r')
plt.ylabel('f(r)')
plt.legend()
plt.ylim(0,1.5)
plt.title('Reconstruction of noisy signal')
plt.savefig('Noise-Reconstruction')
plt.show()
```

## Displacement from the origin

```python
#func=np.exp(-circs.R**2/8**2)
```

```python
orig_signal = I@func
shifted_signal = np.concatenate((orig_signal[:-1],np.flip(orig_signal)))
shifted_signal = np.concatenate((np.ones(20)*np.min(orig_signal),shifted_signal,np.ones(20)*np.min(orig_signal)))
shifted_signal = np.roll(shifted_signal,-2)
shifted_signal = shifted_signal[19+len(ref.r):-20]
shifted_signal = np.flip(shifted_signal)
ext_domain = np.concatenate((-np.flip(ref.r)[:-1],ref.r))
```

```python
plt.plot(ref.r,np.flip(shifted_signal),marker='+',ms=6,label='Offset signal')
plt.plot(ref.r,np.flip(orig_signal),color='black',alpha=0.6,label='Original signal')
plt.xlabel('y')
plt.ylabel('F(y)')
plt.legend()
plt.title('Shifted signal')
plt.savefig('Shifted Signal')
plt.show()
```

```python
recon2 = abel.basex.basex_transform(np.flip(shifted_signal), verbose=True, basis_dir=None, dr = ref.dr, direction='inverse')
plt.plot(ref.r,recon2,marker='+',ms=6,ls='None',label='BASEX reconstruction')
plt.plot(ref.r,inv@shifted_signal,marker='x',ms=6,ls='None',label='Discrete Isoline reconstruction')
plt.plot(ref.r,func,color='black',alpha=0.6,label='Original inverse')
plt.xlabel('r')
plt.ylabel('f(r)')
plt.legend()
plt.title('Reconstruction of shifted signal')
plt.savefig('Shifted-Reconstruction')
plt.show()
```
