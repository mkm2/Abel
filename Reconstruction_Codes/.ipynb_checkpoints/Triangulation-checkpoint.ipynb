```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri
```

## Basic triangulation test

```python
r = [0.1*n for n in range(0,11)]
phi = [np.pi/2/10*n for n in range(0,11)]
data = np.empty((0,2))
for i in range(0,len(r)):
    data_tmp = np.array([r[i]*np.cos(phi),r[i]*np.sin(phi)]).transpose()
    data = np.concatenate((data,data_tmp))
```

```python
triangles = tri.Triangulation(data[:,0], data[:,1])

plt.triplot(triangles)
plt.show()
```

## Area formulas

```python
def tri_area_shoelace(p1,p2,p3):
    area = 0.5*np.abs(p1[0]*p2[1]+p2[0]*p3[1]+p3[0]*p1[1]-p2[0]*p1[1]-p3[0]*p2[1]-p1[0]*p3[1])
    return area
def tri_area_heron(p1,p2,p3):
    a = np.linalg.norm(p1-p2)
    b = np.linalg.norm(p2-p3)
    c = np.linalg.norm(p3-p1)
    s = 0.5*(a+b+c)
    area = np.sqrt(s*(s-a)*(s-b)*(s-c))
    return area
def tri_area_heron_stable(p1,p2,p3):
    a = np.linalg.norm(p1-p2)
    b = np.linalg.norm(p2-p3)
    c = np.linalg.norm(p3-p1)
    C,B,A = np.sort([a,b,c])
    area = 0.25*np.sqrt((A+(B+C))*(C-(A-B))*(C+(A-B))*(A+(B-C)))
    return area
```

```python
for i in range(0,1000000):
    p1 = np.random.uniform(0,1,2)
    p2 = np.random.uniform(0,1,2)
    p3 = np.random.uniform(0,1,2)
    A1 = tri_area_shoelace(p1,p2,p3)
    A2 = tri_area_heron(p1,p2,p3)
    A3 = tri_area_heron_stable(p1,p2,p3)
    diff = np.max([np.abs(A1-A2),np.abs(A1-A3),np.abs(A2-A3)])/np.mean([A1,A2,A3])
```

```python
diff
```

No significant difference. Testing the values from the paper also only revealed a maximum error of $10^{-7}$ in extreme cases.

```python
%%timeit
tri_area_shoelace(p1,p2,p3)
```

```python
%%timeit
tri_area_heron(p1,p2,p3)
```

```python
%%timeit
tri_area_heron_stable(p1,p2,p3)
```

Shoelace method is fastest.
