```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
N = np.array([5,11,21,41,61,81])
M = np.ones(6)*400
```

```python
path = Path.cwd()/'runtimes'
```

```python
start = np.loadtxt(path/'start.txt')
sam = np.loadtxt(path/'sam.txt')
tet = np.loadtxt(path/'tet.txt')
iso = np.loadtxt(path/'iso.txt')
end = np.loadtxt(path/'end.txt')
num = np.loadtxt(path/'num.txt')
```

```python
polycoeffs = np.polyfit(np.log(N),np.log(num),1)
pnum=np.poly1d(polycoeffs)
fit = np.exp(pnum(np.log(np.linspace(1,81))))
plt.loglog(N,num,marker='+',ms=6,ls='None',label='Number of tetrahedra')
plt.loglog(np.linspace(1,81),fit,color='black',alpha=0.6,label=r'$C \cdot N^\alpha$ fit, $\alpha = ${:.2f}'.format(polycoeffs[0]))
plt.xlabel('N')
plt.ylabel('# of tetrahedra')
plt.legend()
plt.title('# of tetrahedra for growing N; M fixed at 400')
plt.show()
```

```python
plt.plot(N,(tet-sam)/60,marker='+',ms=10,ls='None')
plt.xlabel('N')
plt.ylabel('Tetrahedralisation time [min]')
plt.title('# Tetrahedralisation time for growing N; M fixed at 400')
plt.show()
```

```python
plt.plot(N,(end-start)/60,marker='+',ms=10,ls='None')
plt.xlabel('N')
plt.ylabel('Calculation time [min]')
plt.title('Calculation time for growing N; M fixed at 400')
plt.show()
```

```python
polycoeffs = np.polyfit(np.log(N),np.log((end-start)/60),1)
pend=np.poly1d(polycoeffs)
fit = np.exp(pend(np.log(np.linspace(1,81))))
plt.loglog(N,(end-start)/60,marker='+',ms=6,ls='None',label='Calculation time')
plt.loglog(np.linspace(1,81),fit,color='black',alpha=0.6,label=r'$C \cdot N^\alpha$ fit, $\alpha = ${:.2f}'.format(polycoeffs[0]))
plt.xlabel('N')
plt.ylabel('# Calculation time [min]')
plt.legend()
plt.title('Calculation time for growing N; M fixed at 400')
plt.show()
```
